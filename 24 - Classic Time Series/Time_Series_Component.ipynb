# Additive

```python
import pandas as pd
```

```python
df = pd.read_csv('cheese_fred.csv', index_col='date', parse_dates=['date'])
df.head()
```

```python
df.total_cheese_production.plot(figsize=(15, 5))
```

```python
data = df.total_cheese_production
```

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Dekomposisi aditif
result = seasonal_decompose(data, model='additive', period=12)

# Plot hasil dekomposisi
result.plot()
plt.show()

# Akses komponen jika mau
trend = result.trend
seasonal = result.seasonal
residual = result.resid
```

# Multiplicative

```python
df = pd.read_csv('airline.csv', index_col='Date', parse_dates=['Date'])
df.head()
```

```python
df.passengers.plot(figsize=(15, 5))
```

```python
data = df.passengers
```

```python
# Dekomposisi multiplicative
result = seasonal_decompose(data, model='multiplicative', period=12)

# Plot hasil dekomposisi
result.plot()
plt.show()

# Akses komponen kalau mau
trend_component = result.trend
seasonal_component = result.seasonal
residual_component = result.resid
```

# STL

```python
from statsmodels.tsa.seasonal import STL
```

```python
# Dekomposisi STL
stl = STL(data, period=12)
result = stl.fit()

# Plot hasil dekomposisi
result.plot()
plt.show()

# Akses komponen
trend_component = result.trend
seasonal_component = result.seasonal
residual_component = result.resid
```
