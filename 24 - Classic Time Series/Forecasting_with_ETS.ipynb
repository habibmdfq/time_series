# Cheese Production Data

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
df = pd.read_csv('cheese_fred.csv', index_col='date', parse_dates=['date'])
data = df.total_cheese_production
X_train, X_test = data[:-25], data[-25:]
data.plot(figsize=(15, 3))
```

### Training

```python
es = ExponentialSmoothing(X_train, seasonal='add', seasonal_periods=12, use_boxcox=True)
model = es.fit(optimized=True)
```

### Forecast

```python
pred = model.predict(start=len(X_train)-10, end=len(X_train)+50)
```

```python
plt.figure(figsize=(15, 3))
plt.plot(X_train, label='Train')
plt.plot(X_test, label='Test')
plt.plot(pred, label='Prediction')
plt.legend()
```
